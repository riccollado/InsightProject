```python
from scipy.stats import beta
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
```

```python
def mu(a,m,b):
    return (a+4*m+b)/6
```

```python
def sigma(a,m,b):
    return (b-a)/6
```

```python
# Word doc
def alpha_1(a,m,b):
    return ((2*(b+4*m-5*a))/(3*(b-a)))*(1+4*(((m-a)*(b-m))/(b-a)**2))
```

```python
# Classic (webpage)
def alpha_2(a,m,b):
    mean = mu(a,m,b)
    sd = sigma(a,m,b)
    return ((mean-a)/(b-a))*((mean-a)*(b-mean)/(sd**2) - 1)
```

```python
# Vose (webpage)
def alpha_3(a,m,b):
    mean = mu(a,m,b)
    return (mean-a)*(2*m-a-b)/((m-mean)*(b-a))
```

```python
# Priyaanka
def alpha_4(a,m,b):
    return 1 + 4*(m-a)/(b-a)
```

```python
def beta_1(a,m,b):
    return ((2*(5*b-4*m-a))/(3*(b-a)))*(1 + 4*(((m-a)*(b-m))/((b-a)**2)))
```

```python
def beta_2(a,m,b):
    mean = mu(a,m,b)
    return alpha_2(a,m,b)*(b-mean)/(mean-a)
```

```python
def beta_3(a,m,b):
    mean = mu(a,m,b)
    return alpha_3(a,m,b)*(b-mean)/(mean-a)
```

```python
def beta_4(a,m,b):
    return 1 + 4*(b-m)/(b-a)
```

### Test

```python
x = np.linspace(0, 100, 1000)
```

```python
a=0; m=7.5; b=100
loc = a
scale = b-a
```

```python
D1 = beta(alpha_1(a,m,b), beta_1(a,m,b), loc, scale)
D2 = beta(alpha_2(a,m,b), beta_2(a,m,b), loc, scale)
D3 = beta(alpha_3(a,m,b), beta_3(a,m,b), loc, scale)
D4 = beta(alpha_4(a,m,b), beta_4(a,m,b), loc, scale)
```

```python
y1 = D1.pdf(x)
y2 = D2.pdf(x)
y3 = D3.pdf(x)
y4 = D4.pdf(x)
```

```python
plt.plot(x, y1, label='Word')
# plt.plot(x, y2, label='Classic')
# plt.plot(x, y3, label='Vos')
plt.plot(x, y4, label='Priyanka')

plt.legend();
```
